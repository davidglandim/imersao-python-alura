**Aula 2 - Importação**

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df_principal = pd.read_excel("/content/planilha_atualizada_aula2.xlsx", sheet_name = "Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/planilha_atualizada_aula2.xlsx", sheet_name = "Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/planilha_atualizada_aula2.xlsx", sheet_name = "Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/planilha_atualizada_aula2.xlsx", sheet_name = "chatgpt")
df_chatgpt
```

```python
df_principal = df_principal[["Ativo", "Data", "Último (R$)",	"Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns = {"Último (R$)" : "Valor_final", "Var. Dia (%)" : "Var_dia_pct"}).copy()
df_principal
```

```python
df_principal["Var_pct"] = df_principal["Var_dia_pct"] / 100
df_principal["Valor_inicial"] = df_principal["Valor_final"] / (df_principal["Var_pct"] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on = "Ativo", right_on = "Código", how = "left")
df_principal
```

```python
df_principal = df_principal.drop(columns = ["Código"])
df_principal
```

```python
df_principal = df_principal.rename(columns = {"Qtde. Teórica" : "Qtd_teorica"}).copy()
df_principal
```

```python
df_principal["Variacao_rs"] = (df_principal["Valor_final"] - df_principal["Valor_inicial"]) * df_principal["Qtd_teorica"]
df_principal
```

```python
pd.options.display.float_format = "{:.2f}".format
```

```python
df_principal["Qtd_teorica"] = df_principal["Qtd_teorica"].astype(int)
df_principal
```

```python
df_principal["Resultado"] = df_principal["Variacao_rs"].apply(lambda x : "Subiu" if x > 0 else ("Desceu" if x < 0 else "Estável"))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on = "Ativo", right_on = "Ticker", how = "left")
df_principal = df_principal.drop(columns = ["Ticker"])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on = "Nome", right_on = "Empresa", how = "left")
df_principal = df_principal.drop(columns = ["Empresa"])
df_principal
```

```python
df_principal = df_principal.rename(columns = {"Ano de Fundação" : "Ano_de_fundacao", "Idade (em anos)" : "Idade_em_anos"}).copy()
df_principal
```

```python
df_principal["Faixa_etaria"] = df_principal["Idade_em_anos"].apply(lambda x : "Mais de 100 anos" if x > 100 else ("Menos de 50 anos" if x < 50 else "Entre 50 e 100 anos"))
df_principal
```

```python
maior_variacao = df_principal["Variacao_rs"].max()

menor_variacao = df_principal["Variacao_rs"].min()

media = df_principal["Variacao_rs"].mean()

media_subiram = df_principal[df_principal["Resultado"] == "Subiu"]["Variacao_rs"].mean()

media_desceram = df_principal[df_principal["Resultado"] == "Desceu"]["Variacao_rs"].mean()

print(f"Maior:\tR$ {maior_variacao:,.2f}")
print(f"Menor:\tR$ {menor_variacao:,.2f}")
print(f"Média:\tR$ {media:,.2f}")
print(f"Média dos que subiram:\tR$ {media_subiram:,.2f}")
print(f"Média dos que desceram:\tR$ {media_desceram:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal["Resultado"] == "Subiu"]
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby("Segmento")["Variacao_rs"].sum().reset_index()
df_analise_segmento
```

```python
fig = px.pie(df_analise_segmento, values="Variacao_rs", names="Segmento", title="Variação em R$ por segmento")
fig.show()
```

```python
df_analise_saldo = df_principal.groupby("Resultado")["Variacao_rs"].sum().reset_index()
df_analise_saldo
```

```python
# O código do gráfico foi feito com o auxílio do ChatGPT após pesquisar um pouco sobre na documentação do plotly.
# Os próximos comentários foram inseridos pelo ChatGPT. Mantive-os para eu poder me lembrar o que cada parte faz.

# Separar os dados por resultado
subiu = df_analise_saldo[df_analise_saldo["Resultado"] == "Subiu"]["Variacao_rs"]
desceu = df_analise_saldo[df_analise_saldo["Resultado"] == "Desceu"]["Variacao_rs"]
estavel = df_analise_saldo[df_analise_saldo["Resultado"] == "Estavel"]["Variacao_rs"]

# Criar o gráfico de barras
fig = go.Figure()

# Adicionar as barras para "Desceu", "Estavel" e "Subiu"
fig.add_trace(go.Bar(x=["Desceu"], y=desceu, base=0, marker_color="red", name="Desceu"))
fig.add_trace(go.Bar(x=["Estavel"], y=[0], base=0, marker_color="rgba(0,0,0,0)", name="Estavel"))  # Adicionando valor zero para "Estavel" e removendo cor
fig.add_trace(go.Bar(x=["Subiu"], y=subiu, base=0, marker_color="blue", name="Subiu"))

# Atualizar layout do gráfico
fig.update_layout(title="Variação em Reais por Resultado", yaxis=dict(title="Variação em Reais"))

# Mostrar o gráfico
fig.show()
```

```python
df_analise_idade = df_principal.groupby("Faixa_etaria")["Variacao_rs"].sum().reset_index()
df_analise_idade
```

```python
# Código feito utilizando como base o utilizado anteriormente, feito pelo ChatGPT.

mais = df_analise_idade[df_analise_idade["Faixa_etaria"] == "Mais de 100 anos"]["Variacao_rs"]
menos = df_analise_idade[df_analise_idade["Faixa_etaria"] == "Menos de 50 anos"]["Variacao_rs"]
entre = df_analise_idade[df_analise_idade["Faixa_etaria"] == "Entre 50 e 100 anos"]["Variacao_rs"]

fig = go.Figure()

fig.add_trace(go.Bar(x=["Mais de 100 anos"], y=mais, base=0, marker_color="red", name="Mais de 100 anos"))
fig.add_trace(go.Bar(x=["Menos de 50 anos"], y=menos, base=0, marker_color="green", name="Menos de 50 anos"))
fig.add_trace(go.Bar(x=["Entre 50 e 100 anos"], y=entre, base=0, marker_color="blue", name="Entre 50 e 100 anos"))

fig.update_layout(title="Variação em Reais por faixa etária", yaxis=dict(title="Variação em Reais"))

fig.show()
```
